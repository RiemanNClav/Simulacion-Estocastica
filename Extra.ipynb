3 Construir un algoritmo para generar numeros provenientes de una variable aletoria con función de densidad f(x) = 5/12 ((x-1)^4+a)

```python
def f(x):
    y = 5/12*(((x-1)**4)+1)
    return y
def g(x):
    y = 2*((x-1)**3)
    return y
```
